# **5 - Statistical tests**

We want to analyse the correlation in between:
- payment type and payment installment habits
- payment installment habits and payment value

Are there obvious conclusions to be drawn?
We are already assuming that the higher the value, the higher the probability to pay by installments (BUT not a necessity). Let's test that.

Also, we make the hypothesis that there is a positive correlation between the type of payment and the number of installments tied to an order - e.g voucher = 1 payment, credit card = several, etc.

```python
# our full table is df_full_orders

df_full_orders
```

# *Correlation 1*

```python
import pandas as pd
import plotly.graph_objects as go

# We create a contingency table
contingency_table = pd.crosstab(df_full_orders['payment_type'], df_full_orders['payment_installments'])

# Now a heatmap will certainly give us insights on how customers pay and how long they usually need to pay in full

heatmap = go.Figure(data=go.Heatmap(z=contingency_table.values,
                                    x=contingency_table.columns,
                                    y=contingency_table.index,
                                    colorscale='blues'))  # Adjust the colorscale

# Plot labels and title
heatmap.update_layout(title='Correlation Heatmap: Payment Type vs Installments',
                      xaxis_title='Installments',
                      yaxis_title='Payment Type')
heatmap.show()
```
